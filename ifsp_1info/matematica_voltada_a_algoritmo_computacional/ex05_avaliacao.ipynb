```python
a = float(input("Digite o coeficiente a: "))
b = float(input("Digite o coeficiente b: "))
c = float(input("Digite o coeficiente c: "))

delta = (b**2) - (4*a*c)
if delta > 0:
  x1 = (-b + delta**0.5) / (2*a)
  x2 = (-b - delta**0.5) / (2*a)
  print(f"Raízes reais distintas: x1 ={x1}; e x2 ={x2}")
  prova_x1 = a * x1**2 + b * x1 + c
  prova_x2 = a * x2**2 + b * x2 + c
  print(f"Prova para x1: {prova_x1} (deve ser aproximadamente 0)")
  print(f"Prova para x2: {prova_x2} (deve ser aproximadamente 0)")

elif delta == 0:
  x = -b / (2*a)
  print(f"Raiz única (raiz dupla): x = {x}")
  prova_x = a * x**2 + b * x + c
  print(f"Prova para x: {prova_x} (deve ser aproximadamente 0)")

else:
  print("Não existem raízes reais.")
```
